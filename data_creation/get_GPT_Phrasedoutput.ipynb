```python
# --- Step 1: Install and import libraries

# --- Step 3: Upload your input file (with [STYLE_SAMPLE] sections)
print("ğŸ“¤ Upload your input .txt file")

input_filename = "/content/my_style_raw.txt"

# --- Step 4: Read the file
with open(input_filename, "r", encoding="utf-8") as f:
    text = f.read()

# --- Step 5: Split the file by [STYLE_SAMPLE] tags
import re
entries = re.split(r'\[STYLE_SAMPLE\]', text)
entries = [e.strip() for e in entries if e.strip()]

print(f"âœ… Found {len(entries)} style samples to process.")

# --- Step 6: Initialize Gemini model
model = genai.GenerativeModel("gemini-2.5-flash")

# --- Step 7: Rephrase each entry in journalistic tone
rephrased_entries = []

for i, entry in enumerate(entries, 1):
    prompt = f"""
Rephrase the following text in a cumulative journalistic style inspired by Insight IITB (primary),
Stanford Daily, and The New York Times. Keep it concise, factual, and balanced in tone.

Return only the rewritten version without extra commentary.

Original:
{entry}
"""
    response = model.generate_content(prompt)
    new_text = response.text.strip()
    rephrased_entries.append(f"[STYLE_SAMPLE]\n{new_text}\n")
    print(f"âœ… Rephrased entry {i}")

# --- Step 8: Save to a new .txt file
output_filename = "rephrased_style_samples.txt"
with open(output_filename, "w", encoding="utf-8") as f:
    f.write("\n".join(rephrased_entries))

print(f"\nğŸ‰ Done! Output saved as: {output_filename}")


```
